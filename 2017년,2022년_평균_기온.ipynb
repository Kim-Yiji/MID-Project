```python
df1 = pd.read_csv("/content/drive/MyDrive/data/OBS_ASOS_MNH_2017.csv",encoding = 'cp949')
df2 = pd.read_csv("/content/drive/MyDrive/data/OBS_ASOS_MNH_2022.csv",encoding = 'cp949')
df1, df2
```

```python
temp_avg_2017 = df1.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2017
```

```python
temp_avg_2022 = df2.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2022
```

```python
plt.figure(figsize=(50, 10))
plt.xlabel = "날짜"
plt.ylabel = "온도"
plt.plot(temp_avg_2017, label = '2017')
plt.plot(temp_avg_2022, label = '2022')
plt.legend
plt.show()
```
