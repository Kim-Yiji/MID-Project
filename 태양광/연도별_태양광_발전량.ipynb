```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
DATA_PATH = "drive/MyDrive/LikeLion_Data/"
df = pd.read_excel(f"{DATA_PATH}태양광발전량.xlsx")
```

```python
df
```

```python
df.columns
```

```python
cols = ["양수발전", "재생에너지 공급비중", "신에너지 공급비중", "재생에너지 합계", "신에너지 합계"]
df = df.drop(columns=cols)
```

```python
df["신재생에너지 태양광 공급비중"] = df["태양광"].apply(lambda x: float(x.replace(",", ""))) / df["신재생에너지 합계"].apply(lambda x: float(x.replace(",", "")))
df
```

```python
df.to_csv(f"{DATA_PATH}연도별_태양광_비중.csv")
```
