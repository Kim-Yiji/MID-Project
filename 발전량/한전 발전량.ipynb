### 전력 발전량 추이 연-월별

```python
import pandas as pd
df = pd.read_csv("data/발전전력량_추이.csv")
print(df.head())
```

```python
df=df.drop([0,1,2])
print(df.head())
```

```python
print(df.columns)
```

```python
years = ['2017', '2018', '2019', '2020','2021','2022']

rownames = df.columns.tolist()
for i in range(len(years) - 1):
    start_idx = rownames.index(years[i])
    end_idx = rownames.index(years[i + 1])
    

    for j in range(start_idx + 1, end_idx):
        df.drop(columns=[rownames[j]], inplace=True)
```

```python
end_col = '2022'
end_idx = df.columns.get_loc(end_col)


df = df.iloc[:, :end_idx+1]
```

```python
print(df)
```

```python
df.reset_index(drop=True, inplace=True)
df_tras = df.T
print(df_tras)
```

```python
print(df.columns)

# 불필요한 열 삭제 (필요한 경우)
# df.drop(columns=['Unnamed: 0'], inplace=True)

# '월별(1)' 열이 있는지 확인한 후 인덱스 설정
if '월별(1)' in df.columns:
    df.set_index('월별(1)', inplace=True)
else:
    print("Column '월별(1)' not found. Available columns are:", df.columns)

# 인덱스 재설정 (기존 인덱스 번호 제거)
df.reset_index(drop=False, inplace=True)

print(df)
```
