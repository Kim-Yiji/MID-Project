```python
import pandas as pd
data2_2017_2022 = pd.read_csv('data/발전량_지역별.csv', encoding='cp949')
data2_2017_2022
```

```python
data2_2017_2022['전국'] = data2_2017_2022.sum(axis=1)
data2_2017_2022
```

```python
data2_2017_2022_column_통일 = data2_2017_2022.rename(columns={"강원" : "강원도", "경기":"경기도", "경남":"경상남도", "경북":"경상북도", "광주":"광주광역시", "대구": "대구광역시", "전남":"전라남도", "전북":"전라북도", "제주": "제주특별자치도", "충남": "충청남도", "충북": "충청북도", "대전":"대전광역시", "부산" : "부산광역시", "서울":"서울특별시", "세종":"세종특별자치시", "울산":"울산광역시", "인천" : "인천광역시"})
```

```python
data2_2017_2022_column_통일 = data2_2017_2022_column_통일.sort_values("연도").set_index("연도").sort_index(axis=1)
data2_2017_2022_column_통일.to_csv("data/발전량_지역별_컬럼_통일.csv")
```

```python
used_2017_2022 = pd.read_csv('data/사용량_지역별.csv')
used_2017_2022_column_통일 = used_2017_2022.rename(columns={"전체(시도)": "전국"}).set_index("연도").sort_index(axis=1)
used_2017_2022_column_통일.to_csv("data/사용량_지역별_컬럼_통일.csv")
```

```python
produced_used_2017_2022 = data2_2017_2022_column_통일 - used_2017_2022_column_통일
```
