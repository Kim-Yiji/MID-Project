### 시점 처리(연도, 월 추가)

```python
total["연도"] = total['년월'].str[:4]
total["월"] = total['년월'].str[4:6]
total.to_csv('data/사용량_지역별_시간처리.csv', index=False)
```

### 사용량 type 변환

```python
def str_to_int(use):
    return int(use.strip().replace(',', ''))

#str_to_int(total_delete['사용량(kWh)'].iloc[2])
```

```python
total_delete = total.loc[(total['사용량(kWh)'] != '-') & (total['사용량(kWh)']!= '사용량(kWh)')]
total_delete['사용량'] = total_delete['사용량(kWh)'].dropna().apply(str_to_int)
total_delete.to_csv('data/사용량_지역별_시간_사용량처리.csv', index = False)
x = pd.read_csv('data/사용량_지역별_시간_사용량처리.csv')
x
total_delete
```

### groupby

```python
total_groupby = total_delete.groupby(['연도', '시구']).sum().reset_index()
total_groupby_sort = total_groupby[['시구','연도', '사용량']]
total_groupby_sort.to_csv('data/사용량_지역별_처리후_groupby.csv',index = False)
x = pd.read_csv('data/사용량_지역별_처리후_groupby.csv')
x
```
