```python
import pandas as pd
```

```python
df1 = pd.read_csv("C:\practice\practice\지역별\발전량18_22_최종.csv",encoding='cp949')
```

```python
df2 = pd.read_csv("C:\practice\practice\지역별\사용량_지역별_최종_1.csv",encoding='cp949')
```

```python
df1 = df1.reindex(sorted(df1.columns), axis=1)
df2 = df2.reindex(sorted(df2.columns), axis=1)
```

```python
df1
```

```python
df2
```

```python
df1.iloc[:, 1:] = df1.iloc[:, 1:].replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
df2.iloc[:, 1:] = df2.iloc[:, 1:].replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
```

```python
df3 = df1.copy()
df3.iloc[:, 1:] = df1.iloc[:, 1:] - df2.iloc[:, 1:]
```

```python
df3.to_csv("발전량_사용량차이.csv",index=True)
```
