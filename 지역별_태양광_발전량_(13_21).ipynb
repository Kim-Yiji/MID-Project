# 2013-2017 지역별 태양광 발전량 (단위: MWh)

```python
local1317 = pd.read_csv(f"{DATA_PATH}지역별 태양광 발전량(13-17).csv", encoding = "cp949")
local1317
```

```python
local1317.columns
```

```python
local1317["연료원"].unique()
```

```python
local1317["지역세부구분"].unique()
```

```python
local1317.groupby("거래일").nunique().count()
```

```python
local1317["거래일"] = local1317["거래일"].str[:4]
local1317
```

```python
final1317 = pd.pivot_table(
    local1317,
    index = "거래일",
    columns = "지역세부구분",
    values = "전력거래량",
    aggfunc = "sum"
)
final1317
```

```python
total = final1317.sum(axis=1)
total
```

```python
final1317["전국"] = total
final1317
```

# 2018-2021 지역별 태양광 전력거래량 (단위: MWh)

```python
local1821 = pd.read_csv(f"{DATA_PATH}지역별 태양광 전력거래량(18~21년).csv", encoding="cp949")
local1821
```

```python
local1821.columns
```

```python
local1821["지역"].unique()
```

```python
local1821.groupby("거래일자").nunique().count()
```

```python
local1821["거래일자"] = local1821["거래일자"].str[:4]
local1821
```

```python
final1821 = pd.pivot_table(
    local1821,
    index = "거래일자",
    columns = "지역",
    values = "발전량 ",
    aggfunc = "sum"
)
final1821
```

```python
total = final1821.sum(axis=1)
total
```

```python
final1821["전국"] = total
final1821
```

# 테이블 합치기

```python
final1821
```

```python
frames = [final1317, final1821]
final = pd.concat(frames)
final
```
