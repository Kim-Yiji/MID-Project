```python
df = pd.read_csv("/content/drive/MyDrive/data/2017~2022.5~9월 기온.csv",encoding = 'cp949')
df
```

```python
temp_avg_2017 = df.groupby('일시')['평균기온(°C)-2017년'].mean()
temp_avg_2017
```

```python
temp_avg_2018 = df.groupby('일시')['평균기온(°C)-2018년'].mean()
temp_avg_2018
```

```python
temp_avg_2019 = df.groupby('일시')['평균기온(°C)-2019년'].mean()
temp_avg_2019
```

```python
temp_avg_2020 = df.groupby('일시')['평균기온(°C)-2020년'].mean()
temp_avg_2020
```

```python
temp_avg_2021 = df.groupby('일시')['평균기온(°C)-2021년'].mean()
temp_avg_2021
```

```python
temp_avg_2022 = df.groupby('일시')['평균기온(°C)-2022년'].mean()
temp_avg_2022
```

```python
plt.figure(figsize=(50, 10))
plt.xlabel = "날짜"
plt.ylabel = "온도"
plt.plot(temp_avg_2017, label = '2017')
plt.plot(temp_avg_2018, label = '2018')
plt.plot(temp_avg_2019, label = '2019')
plt.plot(temp_avg_2020, label = '2020')
plt.plot(temp_avg_2021, label = '2021')
plt.plot(temp_avg_2022, label = '2022')

def add_trendline(data, label):
    x = np.arange(len(data))  # 날짜에 해당하는 x 값
    z = np.polyfit(x, data, 1)  # 1차 방정식으로 선형 회귀
    p = np.poly1d(z)
    plt.plot(x, p(x), linestyle='--', label=f'{label} 추세선')

add_trendline(temp_avg_2017, '2017')
add_trendline(temp_avg_2018, '2018')
add_trendline(temp_avg_2019, '2019')
add_trendline(temp_avg_2020, '2020')
add_trendline(temp_avg_2021, '2021')
add_trendline(temp_avg_2022, '2022')


plt.legend
plt.show()
```
