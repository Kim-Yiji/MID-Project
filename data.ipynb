# 물가지수

```python
consumer_price_index = pd.read_csv('data/consumer_price_index_2017_2022.csv', encoding='cp949')
```

```python
consumer_price_index = consumer_price_index.rename(columns={'전국': '물가지수'})
```

```python
def float_to_int(num):
    return int(num)
consumer_price_index['연도'] = (consumer_price_index['시점'] // 1).apply(float_to_int)
```

```python
consumer_price_index
```

```python
consumer_price_index['월'] = ((consumer_price_index['시점'] % 1)*100 + 1).apply(float_to_int)
```

```python
consumer_price_index
```

```python
consumer_price_index['전월 물가지수'] = consumer_price_index['물가지수'].shift(1)
```

```python
consumer_price_index['월별 물가 상승률'] = consumer_price_index['물가지수']/consumer_price_index['전월 물가지수']
```

```python
consumer_price_index
```

```python
consumer_price_index.to_csv("data/consumer_price_index_edit.csv", index=False)
```
