```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df2017 = pd.read_csv("/content/drive/MyDrive/data/2017.5~9월 기온.csv",encoding = 'cp949')
df2018 = pd.read_csv("/content/drive/MyDrive/data/2018.05~09월 기온.csv",encoding = 'cp949')
df2019 = pd.read_csv("/content/drive/MyDrive/data/2019.05~09월 기온.csv",encoding = 'cp949')
df2020 = pd.read_csv("/content/drive/MyDrive/data/2020.05~09월 기온.csv",encoding = 'cp949')
df2021 = pd.read_csv("/content/drive/MyDrive/data/2021.05~09월 기온.csv",encoding = 'cp949')
df2022 = pd.read_csv("/content/drive/MyDrive/data/2022.05~09월 기온.csv",encoding = 'cp949')

df2017, df2018, df2019, df2020, df2021, df2022
```

```python
temp_avg_2017 = df2017.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2018 = df2018.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2019 = df2019.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2020 = df2020.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2021 = df2021.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2022 = df2022.groupby('일시')['평균기온(°C)'].mean()
temp_avg_2017, temp_avg_2018, temp_avg_2019, temp_avg_2020, temp_avg_2021, temp_avg_2022
```

```python
temp_avg_2017.count()
```

```python
plt.figure(figsize=(50, 10))
plt.xlabel = "날짜"
plt.ylabel = "온도"
plt.plot(temp_avg_2017, label = '2017')
plt.plot(temp_avg_2018, label = '2018')
plt.plot(temp_avg_2019, label = '2019')
plt.plot(temp_avg_2020, label = '2020')
plt.plot(temp_avg_2021, label = '2021')
plt.plot(temp_avg_2022, label = '2022')
plt.legend
plt.show()
```
