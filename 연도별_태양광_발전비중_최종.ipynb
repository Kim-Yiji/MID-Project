```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
DATA_PATH = "drive/MyDrive/LikeLion_Data/"
df = pd.read_excel(f"{DATA_PATH}태양광발전량.xlsx")
```

```python
df
```

```python
# 단위 GWh
cols = ["양수발전", "재생에너지 공급비중", "신에너지 공급비중", "재생에너지 합계", "신에너지 합계"]
df = df.drop(columns=cols)
```

```python
df
```

```python
df["신재생에너지 태양광 비중"] = (df["태양광"].apply(lambda x: float(x.replace(",", ""))) / df["신재생에너지 합계"].apply(lambda x: float(x.replace(",", ""))))*100
df
```

```python
df["총태양광 비중"] = (df["태양광"].apply(lambda x: float(x.replace(",", ""))) / df["총발전량"].apply(lambda x: float(x.replace(",", ""))))*100
df
```

```python
df["총나머지 비중"] = df["신재생에너지 공급비중"] - df["총태양광 비중"]
df
```

```python
df["총발전량 나머지 비중"] = 100 - df["신재생에너지 공급비중"]
df
```

```python
df["총발전량 비중"] = df["총태양광 비중"] + df["총발전량 나머지 비중"] + df["총나머지 비중"]
df
```

```python
df.to_csv(f"{DATA_PATH}연도별_태양광_비중_최종.csv")
```
