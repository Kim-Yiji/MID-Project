```python
import pandas as pd
DATA_PATH = "data/"
df2 = pd.read_csv("data/발전전력량_추이.csv")
df2
```

```python
filtered_columns = df2.columns[df2.iloc[0].str.contains('총발전량', na=False)]


filtered_df = df2[filtered_columns]
print(filtered_df)
```

```python
# 열 삭제
filtered_df2 = filtered_df.drop(columns=[filtered_df.columns[5], filtered_df.columns[7]])
print(filtered_df2)
```

```python
# 행삭제
filtered_df2 = filtered_df2.drop([0, 1, 2])
filtered_df2.reset_index(drop=True, inplace=True)
print(filtered_df2)
```

```python
filtered_df2.columns=['2017','2018','2019','2020','2021','2022']
print(filtered_df2)
```

```python
filtered_df2.to_csv('발전량_민전총량.csv', header=True)
```
