```python
from google.colab import drive
drive.mount('/content/drive')
DATA_PATH = "/content/drive/MyDrive/LikeLion_Data/"

import pandas as pd
```

```python
local19 = pd.read_excel(DATA_PATH + "지역별 태양광 발전량 19.xls")
local19
```

```python
cols = local19.columns[1:13].tolist()
cols
```

```python
for col in cols:
  local19[col] = local19[col].str.replace(",", "").astype(float)
```

```python
local19.info()
```

```python
local19["총합"] = local19.iloc[:, 1:13].sum(axis=1)
local19
```

```python
final19 = pd.pivot_table(local19, index="구분", values="총합")
final19
```

```python
cols = local20.columns[1:13].tolist()
for col in cols:
  local20[col] = local20[col].str.replace(",","").astype(float)
```

```python
local23 = pd.read_excel(DATA_PATH + "지역별 태양광 발전량 23.xls")
local23
```

```python
cols = local23.columns[1:13].tolist()
for col in cols:
  local23[col] = local23[col].str.replace(",","").astype(float)
local23["총합"] = local23.iloc[:, 1:13].sum(axis=1)
local23
```

```python
years = [2019, 2020, 2021, 2022, 2023]
final = pd.concat([final19, final20, final21, final22, final23], axis=1)
final.columns = years
final = final.transpose()
final = final.reset_index()
```

```python
final.to_csv(DATA_PATH + "지역별 태양광 발전량 (19-23).csv")
```
