```python
df = pd.read_csv("C:\practice\practice\발전량_지역별_시간처리.csv")
df
```

```python
df['사용량'] = pd.to_numeric(df['사용량(kWh)'].str.replace(',', ''), errors='coerce')
df = df[~df['시구'].isin(['황해북도', '미분류'])]
df
```

```python
yms1 = df[['시구', '연도', '사용량']]
yms1
```

```python
counts = yms1['시구'].value_counts()
gangwon_count = counts.get('전체(시도)', 0)
print(f" 개수: {gangwon_count}")
```

```python
yms2 = yms1.groupby(['시구','연도'])['사용량'].sum().reset_index()
yms2
```

```python
yms2.to_csv('2017_2022 사용량 총합.csv',index=False,header=True)
```

```python
경기도 = yms2[yms2['시구'] == '경기도']
outpath = '경기도_groupby.csv'
경기도.to_csv(outpath, index=False,header=True)
```
