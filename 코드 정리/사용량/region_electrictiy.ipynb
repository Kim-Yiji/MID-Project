```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = 'C:/likelion/미드 프로젝트/data/발전량_지역별_시간처리.csv'

df = pd.read_csv(f'{DATA_PATH}')
df
```

```python
# 대전

대전광역시 = df[df["시구"] == "대전광역시"]

cols = ['시구', '연도', '사용량(kWh)']
대전광역시 = 대전광역시.loc[:, cols]

mask = 대전광역시['사용량(kWh)'] != '-'
대전광역시 = 대전광역시[mask]

대전광역시['사용량(kWh)'] = 대전광역시['사용량(kWh)'].str.replace(',', '').str.strip().astype(int)
대전광역시
```

```python
대전광역시.info()
```

```python
대전광역시 = pd.DataFrame(대전광역시.groupby(['연도'])['사용량(kWh)'].sum()).reset_index()
대전광역시.insert(0, '시구', '대전광역시')

대전광역시.to_csv("data3/대전광역시.csv", index=False)
```
